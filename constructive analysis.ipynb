#   Visualization and Interpretation
In 2015, there was increased publicity against the USA police force as people began to register their displeasure about the ugly menance pepertuated by the USA police force. It was observed that officers on duty were shooting indiscrimately and several people were victims of that ugly act.
The data gathered and analysed above was from the 2015-2020 incidents. The data relies primarily on new accounts, social media postings and police reports

```python
year_column = df["year"]
type(year_column)
```

```python
# to check for the year the killing occured most
num_bins = 11
n, bins, patches = plt.hist(year_column, num_bins, alpha=0.9, ec="black")


x_pos = np.arange(len(year_column))
plt.xlabel('year')
plt.ylabel('frequency')
```

# From the graph above, it was observed that the shootings occurred most in 2015, there was a slight decrease in 2016 but that wasn't sustainable as it rose again in 2017. However it is noticeable from the chart that the menance has been greatly curbed in 2020 as there is a significant decrease of almost 60% in the 2020.

```python
# To ascertain the danger mostly affected
gender_column = df["gender"]
type(gender_column)

num_bins = 2
n, bins, patches = plt.hist(gender_column, num_bins, alpha=0.9, ec="black")
```

# Majority of the victims were male as can be seen from the chart. Over 95% of the people that were shot were male and young between the age of 25-29years old.

```python
#to check for the state where the killing occurred most
state_column = df["state"]
type(state_column)


x_pos = np.arange(len(state_column))
plt.xlabel('state')
plt.ylabel('frequency')

num_bins = 60
n, bins, patches = plt.hist(state_column, num_bins, alpha=0.9, ec="black")
```
